```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

cotacao_ibov =  yf.download('^BVSP', data_source='yahoo', start='2020-01-01', end='2020-11-10')
display(cotacao_ibov)
```

```python
cotacao_ibov['Adj Close'].plot(figsize=(15,5))
```

**Retorno do IBOV**

```python
retorno_ibov = cotacao_ibov['Adj Close'][-1]/cotacao_ibov['Adj Close'][0]-1
print('Retorno IBOV: {:.2%}'.format(retorno_ibov))
```

**Médias Móveis**

```python
cotacao_ibov['Adj Close'].plot(figsize=[15,5], label='IBOV')
cotacao_ibov['Adj Close'].rolling(50).mean().plot(label='SMA50')
cotacao_ibov['Adj Close'].rolling(100).mean().plot(label='SMA100')
plt.legend()
plt.show()
```
